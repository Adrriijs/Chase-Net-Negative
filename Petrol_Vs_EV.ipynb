### Import library

```r
library(repr)
library(tidyverse)
```

### Read file

```r
cars <- read_csv("data/vehicle_emission_dataset.csv")
colnames(cars) <- c("VehicleType", "FuelType", "EngineSize", "AgeOfVehicle", "Mileage",
                     "Speed", "Acceleration", "RoadType", "TrafficConditions", "Temperature",
                     "Humidity", "WindSpeed", "AirPressure", "CO2_Emissions", "NOx_Emissions",
                     "PM2.5_Emissions", "VOC_Emissions")

cars
```

### Select and Seperate data

```r
cars <- cars |>
        select(VehicleType, FuelType, CO2_Emissions) |>
        filter(VehicleType == "Car")

filtered_cars <- cars |>
        filter(FuelType == "Electric" | FuelType == "Petrol")

cars_ev <- cars |>
        filter(FuelType == "Electric")

cars_ev

cars_petrol <- cars |>
        filter(FuelType == "Petrol")

cars_petrol
```

### Method 1 : Comparison using boxplot

```r
ggplot(filtered_cars, aes(x = FuelType, y = CO2_Emissions, fill = FuelType)) +
  geom_boxplot() +
  labs(title = "CO2 Emissions: EV vs. Petrol Cars",
       x = "Fuel Type",
       y = "CO2 Emissions (g/km)") +
  theme_minimal()
```

#### Conclusion

The boxplot suggests that **electric and petrol cars have similar CO₂ emission distributions**, which is unexpected since EVs should theoretically produce lower emissions. This could indicate **data issues, misclassification, or the inclusion of hybrid vehicles in the "Electric" category**. Additionally, if the dataset accounts for **indirect emissions** from electricity generation, it may explain the higher-than-expected CO₂ levels for EVs. Further investigation is needed to verify **data accuracy, emission measurement methods, and vehicle classification** to ensure meaningful conclusions. 🚀
